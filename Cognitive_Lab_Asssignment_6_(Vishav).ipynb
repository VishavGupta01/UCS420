Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
- Generate x values from -10 to 10 using np.linspace().
- Compute y values for:
 - y= M⋅x^2
 - y=M⋅sin(x)
- Plot both functions in a single figure:
 - Use different colors and line styles.
 - Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'r-', label=r'$y = Mx^2$')
plt.plot(x, y2, 'b--', label=r'$y = M \sin(x)$')

plt.title("Graphs of y = Mx^2 and y = Msin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)

plt.show()
```

Q.2 Create a dataset of five subjects and scores.
- Convert it into a Pandas DataFrame.
- Plot the scores using a Seaborn bar plot with:
 - Different colors for each bar.
 - Annotations on top of each bar.
 - Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subjects': ['Math', 'Science', 'English', 'History', 'Computer Science'],
    'Scores': [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Subjects', y='Scores', data=df, palette='viridis')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

plt.title("Subject Scores", fontsize=14)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using ``np.random.randn()``.

- Create a 2x2 subplot layout displaying:
 - Line plot showing cumulative sum.
 - Scatter plot with random noise.
- Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)

data = np.random.randn(50)

cumulative_sum = np.cumsum(data)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(cumulative_sum, 'b-', marker='o', label='Cumulative Sum')
axes[0].set_title("Cumulative Sum Line Plot")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Cumulative Sum")
axes[0].legend()
axes[0].grid(True)

axes[1].scatter(range(50), data, color='r', marker='x', label='Random Noise')
axes[1].set_title("Scatter Plot with Random Noise")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Value")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Q 4. Apply ``seaborn`` library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="b", ax=axes[0])
axes[0].set_title("Total Profit per Month")
axes[0].set_xlabel("Month Number")
axes[0].set_ylabel("Total Profit")
axes[0].grid()

for column in df.columns[1:-1]:
    sns.lineplot(x=df["month_number"], y=df[column], label=column, ax=axes[1])
axes[1].set_title("Product Sales per Month")
axes[1].set_xlabel("Month Number")
axes[1].set_ylabel("Sales")
axes[1].legend()
axes[1].grid()

df.set_index("month_number").plot(kind="bar", ax=axes[2], colormap="viridis")
axes[2].set_title("Sales Data")
axes[2].set_xlabel("Month Number")
axes[2].set_ylabel("Values")
axes[2].grid(axis='y')

plt.tight_layout()
plt.show()
```
